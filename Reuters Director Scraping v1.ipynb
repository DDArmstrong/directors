```python
# The URL for all company info is: 'https://www.reuters.com/companies/{ticker}/{section}'

company_dataframe = []

for ticker in ticker_list:
    try:
        response = requests.get('https://www.reuters.com/companies/'+ticker+'/profile')
        doc = bs(response.text, "html.parser")
        company = doc.select_one("div h1:nth-of-type(1)").text
        industry = doc.select_one("div p:nth-of-type(2)").text
        address = doc.find(class_="About-address-AiNm9")
        address = address.find_all("p")
        address1 = []
        for line in address:
            row = line.text
            address1.append(row)
        marketcap = doc.select_one("tbody tr:nth-of-type(10)").text
        company_profile = [ticker,company,industry,address1,marketcap]
        company_dataframe.append(company_profile)
        time.sleep(5)
        
    except:
        f = open("ticker_errors.txt", "a")
        f.write(ticker)
        f.write("\n")
        f.close()
        pass

df = pd.DataFrame(company_dataframe)
df.to_csv('company_info.csv')
    
```
